## **Task 1: Convolution vs Cross-Correlation**

Red cells $\begin{bmatrix} 0 & 1 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 0 \end{bmatrix}$ and given kernel $1/45 \times \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$

* Applying the given kernel to the red cells by cross-correlation, the result will be $\begin{bmatrix} \frac{1}{3} & \frac{13}{45} & \frac{11}{45} \\ \frac{1}{5} & \frac{7}{45} & \frac{1}{9} \\ \frac{1}{15} & \frac{2}{45} & \frac{1}{45} \end{bmatrix}$.


* Applying the given kernel to the red cells by convolution, the result will be $\begin{bmatrix} \frac{1}{9} & \frac{7}{45} & \frac{1}{5} \\ \frac{11}{45} & \frac{13}{45} & \frac{1}{3} \\ \frac{7}{45} & \frac{8}{45} & \frac{1}{5} \end{bmatrix}$.

Both answers show the bottom row similar to its kernel. The convolution result matches the kernel's row, while the cross-correlation result reverses it.



## **Task 2: Tutorial**

```python
# Read 'test_img1.png' and 'test_img2.png'
img_1 = cv2.imread('test_img1.png', cv2.IMREAD_COLOR)
img_2 = cv2.imread('test_img2.png', cv2.IMREAD_COLOR)

# Convert them to grayscale.
img1_gray = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)

# Show them
plt.figure(figsize = [10,5])

plt.subplot(121)
plt.axis('off')
plt.imshow(img1_gray, cmap = 'gray')
plt.title("Image 1", size = 16)

plt.subplot(122)
plt.axis('off')
plt.imshow(img2_gray, cmap = 'gray')
plt.title("Image 2", size = 16)
```

```python
edges_1 = cv2.Canny(img1_gray, threshold1 = 180, threshold2 = 200)
edges_2 = cv2.Canny(img2_gray, threshold1 = 20, threshold2 = 200)

# Show them
plt.figure(figsize = [10,5])

plt.subplot(121)
plt.axis('off')
plt.imshow(edges_1, cmap = 'gray')
plt.title("Image 1", size = 16)

plt.subplot(122)
plt.axis('off')
plt.imshow(edges_2, cmap = 'gray')
plt.title("Image 2", size = 16)
```

```python
# Using cv2.bitwise_and function to AND the two edge images
merged_img = cv2.bitwise_and(edges_1, edges_2)

# Show the result
plt.imshow(merged_img, cmap = 'gray')
```
